Now we want to know the age,year,generation and country with the highest number of suicides...

i feel its neccesary we know so as to look at the reason and cause for that happening

```python
#age with the highest number of suicides(3)
df1.groupby('age')['suicides_no'].sum().nlargest(3)
```

```python
#age with the highest number of suicides(3)
df1.groupby('year')['suicides_no'].sum().nlargest(3)
```

```python
#age with the highest number of suicides(3)
df1.groupby('generation')['suicides_no'].sum().nlargest(3)
```

```python
#age with the highest number of suicides(3)
df1.groupby('country')['suicides_no'].sum().nlargest(3)
```

```python
#Barcharts showing the country with the highest sucides_no  and age with the highest suicides_no
plt.subplot(1,2,1)
df1.groupby('country')['suicides_no'].sum().nlargest(3).plot(kind = 'bar', grid = True)
plt.title('suicides_no(3)')
plt.xlabel('country')
plt.ylabel('suicides_no')
plt.subplot(1,2,2)
df1.groupby('age')['suicides_no'].sum().nlargest(3).plot(kind = 'bar', grid = True,
                                                                       color = 'red')
plt.title('suicides_no(3)')
plt.xlabel('age')
plt.ylabel('suicides_no')
plt.tight_layout()
plt.show()
```

```python
#Barcharts showing the country with the highest sucides_no  and age with the highest suicides_no
plt.subplot(1,2,1)
df1.groupby('generation')['suicides_no'].sum().nlargest(3).plot(kind = 'bar', grid = True)
plt.title('suicides_no(3)')
plt.xlabel('generation')
plt.ylabel('suicides_no')
plt.subplot(1,2,2)
df1.groupby('year')['suicides_no'].sum().nlargest(3).plot(kind = 'bar', grid = True,
                                                                       color = 'red')
plt.title('suicides_no(3)')
plt.xlabel('year')
plt.ylabel('suicides_no')
plt.tight_layout()
plt.show()
```
